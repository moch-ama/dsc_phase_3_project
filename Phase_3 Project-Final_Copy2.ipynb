### Background Information
Traffic stops are encounters between law enforcement and drivers during which the vehicle is stopped for various reasons, such as traffic violations, suspicion of impaired driving, or other potential legal violations. These stops are a common practice for enforcing traffic laws and ensuring public safety.

"Terry stops" refer to stop-and-frisk encounters initiated by law enforcement based on the decision in the Terry v. Ohio case by the United States Supreme Court. In this case, the Court ruled that law enforcement officers can stop and frisk an individual if they have a reasonable suspicion that the person is, was, or will be involved in criminal activity and may be armed and dangerous.

### Column names and their descriptions
The following are the descriptions of columns in the dataset:
<ul>
    <li>-Subject Age Group: Subject Age Group (10 year increments) as reported by the officer.</li>
<li>-Subject ID: Key, generated daily, identifying unique subjects in the dataset using a character to character match of first name and last name. "Null" values indicate an "anonymous" or "unidentified" subject. Subjects of a Terry Stop are not required to present identification.
<li>-GO/SC Num: General Offense or Street Check number, relating the Terry Stop to the parent report. This field may have a one to many relationship in the data.
<li>-Terry Stop ID: Key identifying unique Terry Stop reports.
<li>-Stop Resolution: Resolution of the stop as reported by the officer.
<li>-Weapon Type: Type of weapon, if any, identified during a search or frisk of the subject. Indicates "None" if no weapons was found.
<li>-Officer ID: Key identifying unique officers in the dataset.
<li>-Officer YOB: Year of birth, as reported by the officer.
<li>-Officer Gender: Gender of the officer, as reported by the officer.
<li>-Officer Race: Race of the officer, as reported by the officer.
<li>-Subject Perceived Race: Perceived race of the subject, as reported by the officer.
<li>-Subject Perceived Gender: Perceived gender of the subject, as reported by the officer.
<li>-Reported Date: Date the report was filed in the Records Management System (RMS). Not necessarily the date the stop occurred but generally within 1 day.
<li>-Reported Time: Time the stop was reported in the Records Management System (RMS). Not the time the stop occurred but generally within 10 hours.
<li>-Initial Call Type: Initial classification of the call as assigned by 911.
<li>-Final Call Type: Final classification of the call as assigned by the primary officer closing the event.
<li>-Call Type: How the call was received by the communication center.
<li>-Officer Squad: Functional squad assignment (not budget) of the officer as reported by the Data Analytics Platform (DAP).
<li>-Arrest Flag: Indicator of whether a "physical arrest" was made, of the subject, during the Terry Stop. Does not necessarily reflect a report of an arrest in the Records Management System (RMS).
<li>-Frisk Flag: Indicator of whether a "frisk" was conducted, by the officer, of the subject, during the Terry Stop.
<li>-Precinct: Precinct of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.
<li>-Sector: Sector of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.
<li>-Beat: Beat of the address associated with the underlying Computer Aided Dispatch (CAD) event. Not necessarily where the Terry Stop occurred.</ul>

I will create a function to help in plotting the distributions of this and other various features.

```python
def plot_arrest_distribution(df, category_column):
    # Getting unique values from the specified column
    categories = df[category_column].unique()

    # Num of rows and columns for subplots
    num_rows = 2
    num_cols = int(np.ceil(len(categories) / num_rows))

    # Creating subplots
    fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 8), sharey=True)

    # Flatten the axs array
    axs = axs.flatten()

    # Loop through each category and visualize
    for i, category in enumerate(categories):
        data = df[df[category_column] == category]
        
        arrested_percentage = (data['Stop Resolution'].value_counts(normalize=True) * 100).get('Arrested', 0)
        not_arrested_percentage = 100 - arrested_percentage

        # Visualize a category at a time
        axs[i].pie([arrested_percentage, not_arrested_percentage], labels=['Arrested', 'Not Arrested'], autopct='%1.2f%%', colors=['lightcoral', 'skyblue'])
        axs[i].set_title(category, fontsize=10)

    # Hide remaining empty subplots
    for j in range(len(categories), num_rows * num_cols):
        axs[j].axis('off')

    plt.suptitle(f'Percentage Distribution of Arrested vs Not Arrested by {category_column}', fontsize=16)
    plt.tight_layout()
    plt.show()
```

```python
plot_arrest_distribution(df, 'Officer Gender')
```

To evaluate the performance of the classifiers, I will use confusion matrix. Here I will use True Positives, True Negatives, False Positives and False Negatives. furthermore, I will also use Precision, Recall, Accuracy and F1_score to quantify the performance of classifiers.

<ul>
<li>Precision measures how precise the predictions are.
    <br>precision = Number of true positives / Number of predicted positives
    <br><br/>
<li>Recall indicates what percentage of the classes we're interested in were actually captured by the model.
    <br>Recall = Number of true positives / Number of Actual Total Positives
    <br><br/>
<li>Accuracy is probably the most intuitive metric. 
    <br>Accuracy =  (Number of true positives + true negatives) / Total observartions
    <br><br/>
<li>F1 score represents the Harmonic Mean of Precision and Recall.
    <br> F1 score = 2(Precision x recall) / (Precision + Recall)

Since I will be coming up with confusion matrices for all the models, I will define a function so as to reduce redundancy in coding

```python
def confusion_matrixplot(y_true, y_pred, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Reds):
    '''
    Create a confusion matrix chart showing the accuracy breakdown of the given model.

    Parameters:
    - y_test: True test labels
    - y_pred: Predicted labels
    - classes: Class labels
    - normalize: If True, normalize the data
    - title: Title of the chart
    - cmap: Colors to plot
    '''

    cm = confusion_matrix(y_true, y_pred)
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.ylabel('Test label')
    plt.xlabel('Predicted label')

    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=0)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, f'{cm[i, j]:.2f}' if normalize else cm[i, j],
                 horizontalalignment='center',
                 color='white' if cm[i, j] > thresh else 'black')

    plt.colorbar()
    plt.show()
```

### b). K-Nearest Neighbor (KNN)

Next we will use KNN to predict if an arrest was made or not.
<br>Since KNN arrives at a prediction by voting, we will use odd values for k, to avoid ties and subsequent arbitrary guesswork. By adding this constraint KNN will never be able to evenly split between two classes. 
<br>From here, finding an optimal value of K requires some iterative investigation.

```python
# Defining a function to help get the best k value

def find_best_k(X_train, y_train, X_test, y_test, min_k=1, max_k=49):
    '''This function will determine the best k value to use for knn by selecting the k value
    with the highest overall accuracy score.
    The inputs are:
    X_train, y_train, X_test, y_test: train and test set values
    min_k: minimum value to try for k
    max_k: maximum value to try for k
    '''
    results = [(k, KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train).score(X_test, y_test))
               for k in range(min_k, max_k + 1, 2)]
    
    best_k, best_score = max(results, key=lambda x: x[1])
    
    print("Best k Value: {}".format(best_k))
    print("Accuracy Score: {}".format(best_score))

# Call the function
find_best_k(X_train_resampled, y_train_resampled, X_test_ohe, y_test)
```

```python
# Creating and trainig the KNN classifier n_neighbors=25 (the best value of k)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_resampled, y_train_resampled)

# Make predictions on the test set
y_pred = knn.predict(X_test_ohe)

# Evaluate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'KNN Accuracy: {accuracy * 100:.2f}%')

# Print classification report
print('Classification Report:')
print(classification_report(y_test, y_pred))

# Plot Confusion Matrix
confusion_matrixplot(y_test, y_pred, classes=['Arrested','Not Arrested'], normalize=False, title='KNN Confusion Matrix')
```
